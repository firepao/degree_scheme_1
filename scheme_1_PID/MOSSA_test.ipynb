```python
import numpy as np
from MOSSA import MOSSA

# 定义一个多目标测试函数 (例如 ZDT1 的简化版)
def zdt1(x):
    f1 = x[0]
    g = 1 + 9 * np.sum(x[1:]) / (len(x) - 1)
    f2 = g * (1 - np.sqrt(f1 / g))
    return np.array([f1, f2])

# 参数设置
n_dim = 30
lb = [0] * n_dim
ub = [1] * n_dim

# 初始化并运行
mossa = MOSSA(obj_func=zdt1, n_dim=n_dim, lb=lb, ub=ub, pop_size=50, max_iter=100)
pareto_X, pareto_F = mossa.run()

print("Pareto Front Size:", len(pareto_F))
```
